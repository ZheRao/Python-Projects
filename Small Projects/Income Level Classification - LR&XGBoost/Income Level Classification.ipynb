```python
# adjust loading of the data so that the columns names are not records
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
df.head()
```

```python
# add column names ['age', 'workclass', 'fnlwgt', 'education',
#'education-num', 'marital-status', 'occupation',
#'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
#'hours-per-week', 'native-country', 'income']

df.columns=['age', 'workclass', 'fnlwgt', 'education',
            'education-num', 'marital-status', 'occupation',
            'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
            'hours-per-week', 'native-country', 'income']
df.head()
```

# Null values

```python
# check a general view of null values
df.info()
```

# Non-numerical columns

```python
# check if there is any non-essential columns to be eliminated
df = df.drop("education-num",axis=1)
```

```python
# dummy code the non-numerical columns
df = pd.get_dummies(df)
df.head()
```

```python
# check memory usage
df.info()
```

```python
# determine an appropriate target variable
df=df.drop("income_ <=50K",axis=1)
```

```python
# split X and y
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# Logistic Regression

```python
# import logistic regresion
from sklearn.linear_model import LogisticRegression 
```

```python
# make cross validation available
from sklearn.model_selection import cross_val_score
```

```python
# define a function that takes model and number of folds as input, print accuracy
def cross_validation(model, folds=10):
    scores = cross_val_score(model,X,y,cv=folds)
    print("Accuracies are: \n",np.round(scores,2))
    print("Average accuracy: %.2f%%" % (np.mean(scores)*100))
```

```python
# check the performance of logistic regression
model = LogisticRegression(solver="liblinear")
cross_validation(model)
```

# XGBoost classifier

```python
# import classifier
from xgboost import XGBClassifier
```

```python
# ignore warnings
import warnings
warnings.filterwarnings("ignore")

# limit the number of models to 5, and create cross validation for XGBClassifier
model = XGBClassifier(n_estimators=5,eval_metric="logloss")
cross_validation(model)
```
